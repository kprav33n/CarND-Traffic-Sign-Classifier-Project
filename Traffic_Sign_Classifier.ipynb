**Answer:**

The given image is in RGB format. For image recognition, it was observed that the accuracy was better when we operate on grayscale images. Hence, 

- the image was first converted into grayscale.
- MinMax scaling was employed to scale the values to be in the range of -0.5 to 0.5.

````python
### Generate data additional data (OPTIONAL!)
### and split the data into training/validation/testing sets here.
### Feel free to use as many code cells as needed.

from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

X_train, y_train = shuffle(X_train, y_train, random_state=486)

X_train, X_valid, y_train, y_valid = train_test_split(
    X_train, y_train, test_size=0.2, random_state=992319)

def transforms(img, label):
    rows, cols, _ = img.shape
    m = cv2.getRotationMatrix2D((cols/2, rows/2), 15, 1)
    im_15 = cv2.warpAffine(img, m, (cols, rows))
    m = cv2.getRotationMatrix2D((cols/2, rows/2), -15, 1)
    im_minus15 = cv2.warpAffine(img, m, (cols, rows))
    return np.array([im_15, im_minus15]), np.array([label, label])

X_train_transforms, y_train_transforms = zip(*[transforms(x, y) for x, y in zip(X_train, y_train)])
X_train_transforms = np.array(X_train_transforms)
y_train_transforms = np.array(y_train_transforms)
X_train = np.append(X_train, X_train_transforms.reshape((-1, 32, 32, 3)), axis=0)
y_train = np.append(y_train, y_train_transforms.reshape((-1)), axis=0)

print('Training set:   {} samples, shape: {}'.format(len(X_train), X_train.shape))
print('Validation set: {} samples'.format(len(X_valid)))
print('Testing set:    {} samples'.format(len(X_test)))
````

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

import tensorflow as tf
from tensorflow.contrib.layers import flatten

def weight_variable(shape):
    return tf.Variable(tf.truncated_normal(shape, stddev=0.1))

def bias_variable(shape):
    return tf.constant(0.1, shape=shape)

def model(x, keep_prob):
    num_channels = 3
    n_classes = 43
    conv_kwargs = {
        'strides': [1, 1, 1, 1],
        'padding': 'VALID',
    }
    max_pool_kwargs = {
        'ksize': [1, 2, 2, 1],
        'strides': [1, 2, 2, 1],
        'padding': 'VALID',
    }
    widths = [num_channels, 6, 16, 400, 200, 100, n_classes]
    layer = 1

    c0_W = weight_variable((1, 1, 3, num_channels))
    c0_b = bias_variable((num_channels,))
    c0 = tf.nn.conv2d(x, c0_W, **conv_kwargs) + c0_b
    
    c1_W = weight_variable((5, 5, widths[layer - 1], widths[layer]))
    c1_b = bias_variable((widths[layer],))
    c1 = tf.nn.conv2d(c0, c1_W, **conv_kwargs) + c1_b
    c1 = tf.nn.relu(c1)
    c1 = tf.nn.max_pool(c1, **max_pool_kwargs)
    layer += 1
    
    c2_W = weight_variable((5, 5, widths[layer - 1], widths[layer]))
    c2_b = bias_variable((widths[layer],))
    c2 = tf.nn.conv2d(c1, c2_W, **conv_kwargs) + c2_b
    c2 = tf.nn.relu(c2)
    c2 = tf.nn.max_pool(c2, **max_pool_kwargs)
    layer += 1
    
    fc0 = flatten(c2)
    layer += 1
    
    fc1_W = weight_variable((widths[layer - 1], widths[layer]))
    fc1_b = bias_variable((widths[layer],))
    fc1 = tf.matmul(fc0, fc1_W) + fc1_b
    fc1 = tf.nn.relu(fc1)
    layer += 1

    fc2_W = weight_variable((widths[layer - 1], widths[layer]))
    fc2_b = bias_variable((widths[layer],))
    fc2 = tf.matmul(fc1, fc2_W) + fc2_b
    fc2 = tf.nn.relu(fc2)
    layer += 1
    
    fc3_W = weight_variable((widths[layer - 1], widths[layer]))
    fc3_b = bias_variable((widths[layer],))
    fc3 = tf.matmul(fc2, fc3_W) + fc3_b
    
    return fc3
````

**Answer:**

- A Deep Neural Network that employes ConvNet model was selected for solving this problem, as from a previous exercise it was found that a similar network was good at classifing images to a discrete set of classes.
- Convolutional layers were used in the model to facilitate parameter sharing. This keeps the model simpler, with a manageble number of parameters, making training faster, still provide a reasonable classification accuracy.
- relu was used as the activation function, as it empherically performed better.
- Maxpooling was employed as discretization process to:
  - Reduce dimentionality, thereby train the network faster.
  - Reduce overfitting
- Dropout layer with probability 0.5 while training, and 1 while evaluting was employed to further reduce overfitting.
- After a second layer of convolution, the units are flattened, to feed to two three fully connected layers gradually decreasing in the number of outputs. Each layer was further passed through relu activation.
- Standard hyperparameters with mean 0, and standard deviation 0.1 emphirically proven to provide better learning performance. Hence, those values were used.

````python
### Train your model here.
### Feel free to use as many code cells as needed.

x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)

one_hot_y = tf.one_hot(y, n_classes)
````

````python
rate = 0.001

logits = model(x, keep_prob)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
trainable_vars = tf.trainable_variables()
l2_loss = tf.add_n([tf.nn.l2_loss(v) for v in trainable_vars]) * 0.001
loss_operation = tf.reduce_mean(cross_entropy + l2_loss)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````
